# Opening a file in Py
- file_variable =  open("file_name", "mode")


| Mode | Description | File must exist? | Overwrites? | File Pointer Position |
|-------|--------------|------------------|--------------|------------------------|
| `'r'` | Read (text) mode | `Yes` | No | At the beginning |
| `'w'` | Write (text) mode | No | Yes (creates/truncates file) | At the beginning |
| `'a'` | Append (text) mode | No | No (adds to end) | At the end |
| `'x'` | Exclusive creation mode | No | No (fails if file exists) | At the beginning |
| `'r+'` | Read and write mode | `Yes` | No | At the beginning |
| `'w+'` | Read and write mode (overwrite) | No | Yes (creates/truncates file) | At the beginning |
| `'a+'` | Read and append mode | No | No (adds to end) | At the end |
| `'rb'` | Read (binary) mode | `Yes` | No | At the beginning |
| `'wb'` | Write (binary) mode | No | Yes (creates/truncates file) | At the beginning |
| `'ab'` | Append (binary) mode | No | No (adds to end) | At the end |
| `'rb+'` | Read and write (binary) mode | `Yes` | No | At the beginning |
| `'wb+'` | Read and write (binary, overwrite) | No | Yes (creates/truncates file) | At the beginning |
| `'ab+'` | Read and append (binary) mode | No | No (adds to end) | At the end |



```python
file = open("youtube.txt", "r")
content = file.read()
print(content)
file.close()
```

The above method is not right or rather unsafe way to open a file as you must close file manually using f.close()
- If an error happens before f.close(), file stays open
better way to use with statement
also there is no need for try...finally

```python
with open("youtube.txt", 'r') as file_name:
    content = file_name.read()
# file is automatically closed here
print(content)
```
